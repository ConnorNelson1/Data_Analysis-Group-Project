## Background / Demographics of Chapel Hill

>### Demographics of Chapel Hill

The city of Chapel Hill, North Carolina has a population of 60,988 residents. The majority of the population is between 18 and 24 years old, 31.5%. This is a result of the large student population that occupys the city, with UNC being a centered their as well. Residents under 18 make up 17% of the population, residents 65 years and older make up 10% of the population. The population proportion has only 87.2 men to every 100 women, and the median age is 25.6 years old. Despite this being the most educated city in North Carolina, with 77% of residents holding at least an associates degree, close to 20% of the population living in the city are designated below the federal poverty line.

Chapel Hill has a relatively diverse population within the city limits. 72.53% of Chapel Hill residents are White, with the next largest population group being the Asian Demographic, at 12.66%. The rest of the demographics follow as: Black or African American - 9.74%, Mixed - 3.28%, Other - 1.40%, Native American - 0.37%, Pacific Islander - 0.02%. The largest portion of the population is white, with Asian coming in second, and African American being the third largest population. Regardless, the White population is 8 times larger than the African American Population.

>### Police Arresting Habits: What we know

Policing habits and arresting practices differ from one department to another, and even differ between Sheriff's offices and Police departments. So in order to identify specifics for a specific agency, you would need to compare the data to the demographics and specifics, of the aforementioned area. It is a generally accepted fact that minorities are arrested at higher rates than other non-minority races. The police also tend to arrest the most offenders from the 18-34 age range. Men are arrested more often than females as well, by a very large margin. So the individual in society that would be most likely to be arrested by the police, based on these known trends, would be a black male between the ages of 18 and 34.

>### Dataset Details

The dataset that our group is working with is the arrest data of Chapel Hill, North Carolina from the years 2010 to 2016. Features include the date and time the arrest was made, as well as the address of the arrest. The crime committed is given to us as a classification of a felony or a misdemeanor, a UCR code, and a short description of the crime. The arrestee is described with features of age, race, sex, and ethnicity. The dataset has 18,809 rows and 11 features (columns).

This data is compiled off of arrests made by the police officers from the Chapel Hill police department. The idividuals listed are not necessarily convited of the crimes they are listed as arrested for. This is purely who was arrested by police. This includes anyone that was arrested, booked, and released later on in the evening and anyone arrested that might be not be guilty of the crime they allegedly committed. This does not measure the crime rate, as any crime that went unreproted, not suspect was arrested, or the perpatrator was dead or killed before they could be apprehended, were not measured in this study. Only the arrests made, and the corresponding offenses justifying arrest.

```python
# code for changing the ucr_code feature to be coded as either Violent Crime, Property Crime, or Narcotic Crime
# based on the information above

pam_dict1 = {x:'Violent Crime' for x in list(set(pam.ucr_code))
             if x in [110,640,200,300,410,411,510,810,811,820,1530
                      ,1550,1590,1710,1720,1730,1770,1780,2030,2090,2100,2212,
                      2214,2222,2223,2230,2240,2290,2410,2430,2440,2450,2490,
                      2520,2570,2590,2620,2630,2640,2650,2670,2680,2690,4010]}
pam_dict2 = {x:'Property Crime' for x in list(set(pam.ucr_code)) 
             if x in[620,630,670,690,710,900,1015,1027,1110,1120,1140,1150,1170,1180,1190,1210,1290,1310,1320,
                      1330,1400,2020]}
pam_dict3 = {x:'Narcotic Crime' for x in list (set(pam.ucr_code))
             if x in [1810,1820,1834,1890]}

pam_dict1.update(pam_dict2)

pam_dict1.update(pam_dict3)

pam_2 = pam.copy()   
    
pam_2.ucr_code = pam_2.ucr_code.map(pam_dict1) 
```

```python
set(pam_2.ucr_code)
```

```python
# code for changing the ucr_code to be either 0 (Violent Crime), 1 (Property Crime), or 2 (Narcotic Crime)

pam_2_dict4 = {x: 0 for x in list(set(pam_2.ucr_code))
             if x in ['Violent Crime']}
pam_2_dict5 = {x: 1 for x in list(set(pam_2.ucr_code)) 
             if x in['Property Crime']}
pam_2_dict6 = {x: 2 for x in list (set(pam_2.ucr_code))
             if x in ['Narcotic Crime']}

pam_2_dict4.update(pam_2_dict5)

pam_2_dict4.update(pam_2_dict6) 

pam_3 = pam_2.copy()   
    
pam_3.ucr_code = pam_3.ucr_code.map(pam_2_dict4)
```

```python
set(pam.arrestee_s_race)
```

```python
# code for one-hot-encoding the race feature to be numerical rather than categorical 

def onehot(df, feature):

    pamv = list(set(df[feature])) 
    
    for c in df.index: 
        t = df.loc[c, feature]
        
        for test in pamv:
            if pa.isna(test): 
                if pa.isna(t):
                    df.loc[c, '{}_nan'.format(feature)] = 1
                else:
                    df.loc[c, '{}_nan'.format(feature)] = 0
            else:
                if t == test:
                    df.loc[c, '{}_{}'.format(feature, test)] = 1  
                else:
                    df.loc[c, '{}_{}'.format(feature, test)] = 0 
            
    return df.drop(feature, axis=1)
```

```python
pam_hot1 = onehot(pam_3, 'arrestee_s_race')
```

```python
pam_hot1['arrestee_s_age']=pam_hot1['arrestee_s_age'].astype('float')
pam_hot1['arrestee_s_race_A']=pam_hot1['arrestee_s_race_A'].astype('float')
pam_hot1['arrestee_s_race_B']=pam_hot1['arrestee_s_race_B'].astype('float')
pam_hot1['arrestee_s_race_O']=pam_hot1['arrestee_s_race_O'].astype('float')
pam_hot1['arrestee_s_race_H']=pam_hot1['arrestee_s_race_H'].astype('float')
pam_hot1['arrestee_s_race_I']=pam_hot1['arrestee_s_race_I'].astype('float')
pam_hot1['arrestee_s_race_W']=pam_hot1['arrestee_s_race_W'].astype('float')
pam_hot1['arrestee_s_race_U']=pam_hot1['arrestee_s_race_U'].astype('float')
pam_hot1['arrestee_s_race_nan']=pam_hot1['arrestee_s_race_nan'].astype('float')
```

```python
pam_hot1 = pam_hot1.dropna(subset=['arrestee_s_age', 'ucr_code'])
```

### Linear and Quadratic Discriminant Analysis: 

##### Beginning of linear discriminant analysis: 

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
pam_3 = pam_3.dropna(subset=['arrestee_s_age', 'ucr_code'])
```

```python
X = np.atleast_2d(np.array(pam_3['arrestee_s_age'])).transpose()
y = np.array(pam_3['ucr_code'])
```

```python
n = X.shape[0]
test = int(0.20*n)            
perm = rn.permutation(n)   
X = X[perm]
y = y[perm]
X_test = X[:test]       
y_test = y[:test]
X_train = X[test:]     
y_train = y[test:]
```

```python
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train);
print('Training Error: {}'.format(clf.score(X_train, y_train)))
print('Testing Error: {}'.format(clf.score(X_test, y_test)))
```

##### Checking performance with bootstraps: 

```python
num = 60 
Xboot = [0]*num
yboot = [0]*num
```

```python
n = X.shape[0]
for s in range(num):
    sample = np.random.randint(0, n, n)  
    Xboot[s] = X[sample, :]
    yboot[s] = y[sample]
```

```python
test = int(0.20*n)
maxk = 20
labels = [ ['train_{}'.format(x), 'test_{}'.format(x)] for x in range(1, maxk+1)]
labels = [x for sublist in labels for x in sublist]
```

```python
Error = pa.DataFrame(  np.array([0]*num*2*maxk).reshape(num, 2*maxk), columns = labels)

for s in range(num):
    X_test = Xboot[s][:test]
    y_test = yboot[s][:test]
    X_train = Xboot[s][test:]
    y_train = yboot[s][test:]
    
    for k in range(maxk):
        clf = LinearDiscriminantAnalysis()
        clf.fit(X_train, y_train)
        Error.iloc[s, 2*k] = clf.score(X_train, y_train) 
        Error.iloc[s, 2*k+1] = clf.score(X_test, y_test)
      
Error_mean = [0]*2*maxk
Error_std = [0]*2*maxk
for c in range(2*maxk):
    Error_mean[c] = np.mean(Error.iloc[:num, c])
    Error_std[c] = np.std(Error.iloc[:num, c])
    
Error = Error.append( pa.Series(Error_mean, index=Error.columns), ignore_index=True)
Error = Error.append( pa.Series(Error_std, index=Error.columns), ignore_index=True)
Error.index = list(range(num)) + ['Mean', 'Standard Deviation']
Error.tail()
```
